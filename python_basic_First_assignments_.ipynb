       Question 2:-What is an Interpreter in Python?

   ANS:-  An interpreter in Python is a program that reads and executes Python code line by line. It translates the high-level Python code you write (which humans can understand) into machine code.

How it works:

1. You write Python code (.py file).


2. The Python interpreter reads one line of code at a time.


3. It converts that line to machine instructions and immediately executes it.


4. This happens repeatedly until the program finishes.



Example:

print("Hello, World!")

When you run this code using the interpreter (e.g., by typing python myfile.py in the terminal), the interpreter will:

Read print("Hello, World!")

Convert it to machine instructions

Execute it to show Hello, World! on your screen.


Types of Python interpreters:

CPython (default and most common, written in C)

PyPy (faster, uses Just-In-Time compilation)

Jython (Python interpreter for Java)

IronPython (Python interpreter for .NET)

    Question 5:-What is mutability in python?

 Ans:-Mutability in Python refers to whether an object can be changed after it is created.

If an object can be changed, it is called mutable.

If an object cannot be changed, it is called immutable.

Why does it matter?

Mutability affects how objects behave when you assign them to variables, pass them to functions, or copy them.

Mutable objects can be modified in place.

Immutable objects require creating a new object if you want to change the value.


    Example:

    Mutable Objects	Immutable Objects

    list	int
    dict	float
    set	str (string)
    bytearray	tuple
    user-defined classes (usually mutable)	frozenset

       Example of mutable object (list):

     my_list = [1, 2, 3]
     print(my_list)      # Output: [1, 2, 3]

    my_list.append(4)   # Modifying the list in place
    print(my_list)      # Output: [1, 2, 3, 4]

    my_list was changed — it is mutable.


     Example of immutable object (str):

    my_str = "hello"
    print(my_str)       # Output: hello
    my_str = "Hello"
    print(my_str)


    Question 6:-Why are list mutuable but tuple is immutable?

   Ans:-Good question — it gets to the heart of how Python is designed. Let's break it down clearly:

    List is mutable

A list in Python is defined like this:

my_list = [1, 2, 3]

Mutable means you can:

Change existing elements → my_list[0] = 10

Add new elements → my_list.append(4)

Remove elements → my_list.pop()


    Why is it mutable?

Lists are designed to be flexible and dynamic.

Internally, Python stores lists as arrays of object references — this allows efficient resizing and element updates.

Many use-cases (such as collecting data, modifying elements) need this flexibility.

    Tuple is immutable

A tuple is defined like this:

my_tuple = (1, 2, 3)

Immutable means you cannot:

Change any element → my_tuple[0] = 10 →  error

Add or remove elements after creation → no .append(), no .pop()


    Why is it immutable?

Tuples are often used to represent fixed collections of data (coordinates, settings, dictionary keys, etc.).

Once created, the tuple guarantees that its content won't change — this helps with:

Safety: Code can trust that the data stays constant.

Hashability: Immutable objects like tuples can be used as dictionary keys or set elements.

Performance: Python can optimize storage of immutable objects more effectively.

    Question 7:- What is the difference between"==" and "is" in python?

Ans:-Good question! In Python, == and is are both comparison operators, but they are used for different purposes:

1. == → Equality Operator

It checks whether the values of two objects are equal.

It can be overridden by defining _eq_() in a class.

    Example:

    a = [1, 2, 3]
    b = [1, 2, 3]
    print(a == b)  # True → because the lists have the same conten
    
2. is → Identity Operator

It checks whether two variables refer to the same object in memory (same identity).

It cannot be overridden.

    Example:

    a = [1, 2, 3]
    b = [1, 2, 3]
    print(a is b)  # False → because a and b are different objects, even though their contents are equal

    c = a
    print(a is c)  # True → because c and a refer to the same object

    Question 9:-What is typecasting in python?

Ans:-Typecasting in Python refers to converting one data type to another. It is useful when you want to perform operations that require values to be in a specific type (e.g., converting a string to an integer so you can perform arithmetic).

There are two main types of typecasting in Python:

    ⿡ Explicit Typecasting (Manual conversion)

You manually convert one type to another using built-in functions:

Function	Converts to

int()	Integer
float()	Floating-point number
str()	String
bool()	Boolean
list()	List
tuple()	Tuple
set()	Set


    Example:

    x = "123"
    y = int(x)      # Converts string to integer
    print(y + 10)   # Output: 133

    a = 5
    b = str(a)      # Converts integer to string
    print(b + " apples")  # Output: "5 apples"


    ⿢ Implicit Typecasting (Automatic conversion)

Python automatically converts one data type to another when necessary, without explicit instruction from the programmer. It usually happens when performing operations between different types.

    Example:

    x = 10      # Integer
    y = 2.5     # Float
    z = x + y   # Integer + Float -> Float
    print(z)    # Output: 12.5

Here, Python implicitly converts x from int to float during the addition

    Question 12:-How does the elif statement work?

Ans:-

The elif statement in Python stands for:

 "else if"

It is used when you want to check multiple conditions, one after another.
It is part of the if-elif-else ladder.

    How it works:

     The program checks the first if condition:

     If it is True, that block runs, and the rest (elif, else) are skipped.

     If the first if is False, it checks the elif condition:

     If that elif is True, its block runs, and the rest are skipped.


     If none of the if or elif conditions are True, the else block (if provided) runs.

    Example:

    number = 0

    if number > 0:
    print("Positive number")
    elif number == 0:
    print("Zero")
    else:
    print("Negative number")

    Output:

    Zero

    Flow of this example:

1. number > 0 → False → skip this block


2. number == 0 → True → run this block ("Zero")


3. else → skipped (since an earlier condition was True)

    Why use elif?

Without elif: you'd have to write multiple separate if statements, which is less efficient and harder to read.

With elif: once one condition is met, the rest are skipped — it’s faster and clearer.


    Question 13:-What is the difference between for and while loop?

Ans:-The difference between for and while loops explained below:-

1. Purpose



For loop: Used when you know the number of iterations or want to iterate over a sequence.

While loop: Used when you want to repeat based on a condition that may change during execution.


Example:

# For loop example: print numbers 1 to 5
    for i in range(1, 6):
    print(i)

# While loop example: print numbers 1 to 5
    i = 1
    while i <= 5:
    print(i)
    i += 1


2. Iteration Control



For loop: Automatically handles loop variable and iterations over a sequence.

While loop: Requires manual update of the loop variable inside the loop.



3. Condition Checking



For loop: Iterates over a predefined sequence or range, condition is implicit.

While loop: Condition is explicitly checked before each iteration.


4. Risk of Infinite Loop



For loop: Less likely to cause infinite loops because the sequence is fixed.

While loop: More prone to infinite loops if the condition is never met or the loop variable is not updated.


5. Common Usage



For loop: When iterating over items (list, string, range).

While loop: When looping until a condition changes, like waiting for user input or event.
